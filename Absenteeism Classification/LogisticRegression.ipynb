```python
import pandas as pd
import numpy as np
```

## Load Data

```python
data_preprocessed=pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data_preprocessed.head()
```

## Create Target 

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets=np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism']=targets
```

```python
data_preprocessed
```

```python
data_with_targets=data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets
```

## Select the inputs for regression

```python
data_with_targets.shape
```

```python
data_with_targets.iloc[:,0:14]
```

```python
data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs=pd.concat([data_with_targets.iloc[:,4:11],data_with_targets.iloc[:,12:-1]],axis=1)
unscaled_inputs
```

```python
unscaled_inputs.columns.values
```

## Data Standardization

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler=StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs=absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs.shape
```

```python
scaled_inputs=pd.DataFrame(scaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.columns=['Month Value', 'Day of the week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Children', 'Pets']
```

```python
scaled_inputs
```

```python
scaled_inputs=pd.concat([scaled_inputs,data_with_targets.iloc[:,:4],data_with_targets.iloc[:,11:12]],axis=1)
```

```python
scaled_inputs
```

```python
column_names=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value','Day of the week', 
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets']
scaled_inputs=scaled_inputs[column_names]
```

```python
scaled_inputs
```

## Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,train_size=0.8)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

### Training the model

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_name=scaled_inputs.columns.values
```

```python
summary_table=pd.DataFrame(columns=['Feature name'],data=feature_name)
```

```python
summary_table
```

```python
summary_table['Coefficients']=np.transpose(reg.coef_)
```

```python
summary_table.index=summary_table.index+1
summary_table.loc[0]=['Intercept',reg.intercept_[0]]
summary_table=summary_table.sort_index()
summary_table
```

## Interpreting Coefficients

```python
summary_table['Odds Ratio']=np.exp(summary_table.Coefficients)
```

```python
summary_table
```

summary_table.sort_values('Odds Ratio',ascending=False)

## Testing

```python
reg.score(x_test,y_test)
```

```python
predicted_proba=reg.predict_proba(x_test)
predicted_proba
```

```python
predicted_proba[:,1]
```

```python
probabilities=predicted_proba[:,1]
```

```python
x_test
```

```python
predictions=reg.predict(x_test)
```

```python
predictions
```

```python
x_test_df=x_test
x_test_df.update(unscaled_inputs)
```

```python
x_test_df
```

```python
predictions_df=x_test_df
predictions_df['probabilities']=probabilities
predictions_df['predictions']=predictions
```

```python
predictions_df
```
